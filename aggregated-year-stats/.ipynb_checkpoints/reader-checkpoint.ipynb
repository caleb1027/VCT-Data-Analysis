```python
class Player():
    def __init__(self, arr):
        self.name = arr[0]
        self.team = arr[1]
        self.rounds = arr[2].item()
        self.acs = arr[3].item()
        self.kpr =arr[4].item()
        self.apr = arr[5].item()
        self.kd = arr[6].item()
        self.kast =arr[7].item()
        self.adr =arr[8].item()
        self.hsp =arr[9].item()
        self.fkpr =arr[10].item()
        self.fdpr =arr[11].item()
        self.fkfd = arr[12].item()
        self.clutch = arr[13].item()
    def set_stats(self, arr):
        self.rounds = arr[2].item()
        self.acs = round(((self.acs + arr[3])/2), 2).item()
        self.kpr = round(((self.kpr + arr[4])/2), 2).item()
        self.apr = round(((self.apr + arr[5])/2), 2).item()
        self.kd = round(((self.kd + arr[6])/2), 2).item()
        self.kast = round(((self.kast + arr[7])/2), 2).item()
        self.adr = round(((self.adr + arr[8])/2), 2).item()
        self.hsp = round(((self.hsp + arr[9])/2), 2).item()
        self.fkpr = round(((self.fkpr + arr[10])/2), 2).item()
        self.fdpr = round(((self.fdpr + arr[11])/2), 2).item()
        self.fkfd = round(((self.fkfd + arr[12])/2), 2).item()
        self.clutch = round(((self.clutch + arr[13])/2), 2).item()
        
    
  
```

```python
player_names =[]
player_list =[]
```

```python
for x, i in enumerate(df.index):
    player_list.append(Player(df.iloc[i].to_numpy()))
    player_names.append(df.iloc[i]['Name'])
    
for x, i in enumerate(df1.index):
    if df1.iloc[i]['Name'] in player_names:

        index = player_names.index(df1.iloc[i]['Name'])
        player_list[index].set_stats(df1.iloc[i].to_numpy())
    else:
        player_list.append(Player(df1.iloc[i].to_numpy()))
        player_names.append(df1.iloc[i]['Name'])
```
